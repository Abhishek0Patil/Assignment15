Matplotlib is a comprehensive library in Python for creating static, animated, and interactive visualizations. It provides a wide range of plotting functions and tools for creating various types of plots and graphs. It is widely used in fields such as data science, machine learning, scientific computing, engineering, and more.
Five plots that can be ploted using Pyplot module are 
1. Line plot
2. Scatter plot
3. Histogram
4. Bar graph
5. Pie chart

Scatter plots are used to display the relationship between two continuous variables. Each data point is represented as a marker on the plot, with the position determined by the values of the variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x1,y1,'r')
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y,'g')
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x2,y2,'y')
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x3,y3,'b')
plt.show()
```

A bar plot is a type of chart that represents categorical data with rectangular bars. 
Bar plots are used to compare the values of different categories or groups. They display the value of each category as a bar, with the height or length of the bar proportional to the value.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit,color='red')
plt.show()
```

A box plot is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a concise way to visualize the central tendency, variability, and skewness of the data distribution, as well as identify potential outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True, vert=False)
plt.yticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Values')
plt.title('Box Plot of Two Datasets')
plt.grid(True)
plt.show()
```
